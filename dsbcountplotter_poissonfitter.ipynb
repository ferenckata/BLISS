```python
breaks = {}
with open('./macs/pooledchr1.bed','r') as bf:
    for line in bf:
        dsbcount = int(line.strip().split()[3])
        if dsbcount not in breaks.keys():
            breaks[dsbcount] = 1
        else:
            breaks[dsbcount] += 1
```

```python
lists = sorted(breaks.items())
x,y = zip(*lists)
print(x,y)
```

```python
def poisson(k, lamb):
    return (lamb**k/factorial(k)) * np.exp(-lamb)
```

```python
parameters, cov_matrix = curve_fit(poisson, x, y)
x_plot = np.linspace(0, 20, 1000)

plt.plot(x_plot, poisson(x_plot, *parameters), 'r-', lw=2)
plt.show()
```
